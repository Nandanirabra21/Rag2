```python
# Import required modules
from langchain.document_loaders import PyMuPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
import os
import spacy
from langchain.schema import Document

# Load SpaCy model for NLP
nlp = spacy.load('en_core_web_sm')

# Load environment variables from .env file
load_dotenv()

# Ensure the API key is set
openai_api_key = os.getenv('OPENAI_API_KEY')
if not openai_api_key:
    raise ValueError("OpenAI API key is not set. Please check your .env file.")

# Load your documents (PDFs)
documents_paths = ["/content/Fin12.pdf", "/content/Fin13.pdf", "/content/Fin14.pdf"]
documents = []

for path in documents_paths:
    loader = PyMuPDFLoader(path)  # Use PyMuPDFLoader for PDFs
    documents.extend(loader.load())  # Load and append each document's content

# Apply SpaCy for semantic chunking
semantic_chunks = []
for i, doc in enumerate(documents):
    spacy_doc = nlp(doc.page_content)  # Process text with SpaCy
    print(f"Document {i+1} - Semantic Chunking:")
    chunk_index = 0
    for sentence in spacy_doc.sents:   # Use sentence boundaries as chunks
        chunk_text = sentence.text.strip()
        if chunk_text:  # Avoid empty chunks
            semantic_chunks.append(chunk_text)
            print(f"  Chunk {chunk_index}: {chunk_text[:100]}...")  # Print the start of the chunk for preview
            chunk_index += 1

# Create Document objects for the chunks
document_objects = [Document(page_content=chunk) for chunk in semantic_chunks]

# Further split the text into chunks using RecursiveCharacterTextSplitter if necessary
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
texts = text_splitter.split_documents(document_objects)

# Print the number of chunks generated
print(f"Number of chunks created after splitting: {len(texts)}")

# Print the first few chunks for preview
for i, chunk in enumerate(texts[:5]):
    print(f"Chunk {i}: {chunk.page_content[:100]}...")  # Print the start of the chunk for preview

# Initialize OpenAI embeddings with the API key
embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)

# Generate embeddings for the first 5 chunks
doc_vectors = embeddings.embed_documents([text.page_content for text in texts[:5]])

# Print the dimensionality of the first vector to confirm
print(f"Dimensionality of first vector: {len(doc_vectors[0])}")
```
